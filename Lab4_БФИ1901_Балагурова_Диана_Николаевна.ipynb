# Лабораторная работа 4. Реализация стека/дека.
Реализовать следующие структуры данных:
● Стек (stack)
● Дек (двусторонняя очередь, deque)

```python
class Stack:
	def __init__(self):
		self.items = []

	def push(self, item):
		self.items.append(item)

	def pop(self):
		return self.items.pop()

	def get(self):
		return self.items[-1]

	def empty(self):
		return self.items == []

	def __str__(self):
		return ','.join(map(str, self.items))

	def __repr__(self):
		return ','.join(map(str, self.items))


class Deque:
	def __init__(self, items = None):
		if items is None:
			self.items = []
		else:
			self.items = items

	def push_bottom(self, item):
		self.items.append(item)

	def push_top(self, item):
		self.items.insert(0, item)

	def pop_bottom(self):
		return self.items.pop()

	def pop_top(self):
		return self.items.pop(0)

	def get_bottom(self):
		return self.items[-1]

	def get_top(self):
		return self.items[0]

	def empty(self):
		return self.items == []

	def __str__(self):
		return ','.join(map(str, self.items))

	def __repr__(self):
		return ','.join(map(str, self.items))


def z1():
	print('___1___')

#	filename = 'books.txt'
#	f = open(filename, "r", encoding='utf-8')
#	books = f.readlines()
	books = ['«Властелин колец» — Дж. Р. Р. Толкиен', 
'«Гордость и предубеждение» — Джейн Остин'
'«Тёмные начала» — Филип Пулман',
'«Автостопом по галактике» — Дуглас Адамс',
'«Гарри Поттер и Кубок огня» — Джоан Роулинг',
'«Убить пересмешника» — Харпер Ли',
'«Винни Пух» — Алан Александр Милн']
	print(books)

	deq1, deq2 = Deque(), Deque()

	for book in books:
		if deq1.empty():
			deq2.push_bottom(book)
		else:
			book_added = False
			while not deq1.empty():
				b = deq1.pop_top()
				if not book_added and book < b:
					deq2.push_bottom(book)
					book_added = True
				deq2.push_bottom(b)
			if not book_added:
				deq2.push_bottom(book)
		deq1, deq2 = deq2, deq1
	print(deq1.items)


def z2():
	print('___2___')
	shifr_deq = Deque(['%', '#', ',', 'l', 'k', 'g', 'o', 'e'])
	shifr_str = 'k,%lkkl#g'
	print(shifr_str)
	clean_str = ''
	for sym in shifr_str:
		while True:
			s = shifr_deq.pop_top()
			shifr_deq.push_bottom(s)
			if s == sym:
				shifr_deq.push_bottom(shifr_deq.pop_top())
				s1 = shifr_deq.pop_top()
				shifr_deq.push_bottom(s1)
				clean_str += s1
				break
	print(clean_str)


def z3(num):
	print('___3___')
	s1 = Stack()
	s2 = Stack()
	s3 = Stack()
	for i in range(1, num + 1):
		s1.push(num - i + 1)

	print('1: ' + str(s1))
	print('2: ' + str(s2))
	print('3: ' + str(s3))

	def solve(numbers, s1, s2, s3, s1_name, s2_name, s3_name):
		if numbers == 1:
			print('Move disk 1 from {} to {}'.format(s1_name, s3_name))
			s3.push(s1.pop())
			return
		solve(numbers - 1, s1, s3, s2, s1_name, s3_name, s2_name)
		print('Move disk {} from {} to {}'.format(numbers, s1_name, s3_name))
		s3.push(s1.pop())
		solve(numbers - 1, s2, s1, s3, s2_name, s1_name, s3_name)

	solve(num, s1, s2, s3, '1', '2', '3')
	print('1: ' + str(s1))
	print('2: ' + str(s2))
	print('3: ' + str(s3))


def z4():
	print('___4___')
#	filename = 'program.txt'
#	f = open(filename, "r", encoding='utf-8')
#	strings = f.readlines()
	strings = ['()()',
	           '(((()())()))',
	           '((()()))((()()))']

	st = Stack()
	for s in strings:
		for b in s:
			if b == '(':
				st.push(b)
			elif b == ')':
				if st.empty():
					print('Imbalance!')
					return
				else:
					st.pop()
	if st.empty():
		print('Balance!')
	else:
		print('Imbalance!')


def z5():
	print('___5___')
	#	filename = 'program.txt'
	#	f = open(filename, "r", encoding='utf-8')
	#	strings = f.readlines()
	strings = ['()()',
	           '(((()())()))',
	           '((()()))((()()))']

	st = Deque()
	for s in strings:
		for b in s:
			if b == '(':
				st.push_top(b)
			elif b == ')':
				if st.empty():
					print('Imbalance!')
					return
				else:
					st.pop_top()
	if st.empty():
		print('Balance!')
	else:
		print('Imbalance!')


def z6():
	print('___6___')
#	filename = 'symbols.txt'
#	f = open(filename, "r", encoding='utf-8')
#	strings = f.readlines()
	strings = ['Procedure SortArray(n : Integer; Var NQ : Array[1..n] of Real);',
	           'Var k: Real; i,j : Integer;',
	           'Begin for i:=1 to n-1 do']

	st_c, st_b, st_o  = Stack(), Stack(), Stack()
	for s in strings:
		for b in s:
			if b.isdigit():
				st_c.push(b)
			elif b.isalpha():
				st_b.push(b)
			elif b.isprintable():
				st_o.push(b)
	print(str(st_c))
	print(str(st_b))
	print(str(st_o))


def z7():
	print('___7___')
#	filename = 'chisla.txt'
#	f = open(filename, "r", encoding='utf-8')
#	strings = f.readlines()
	strings = [-1, 2, -4, 6, -7, -5, 12, 9]

	dec_o, dec_p = Deque(), Deque()
	for s in strings:
		c = int(s)
		if c > 0:
			dec_p.push_bottom(c)
		else:
			dec_o.push_bottom(c)
	print(str(dec_o))
	print(str(dec_p))


def z8():
	print('___8___')
#	filename = 'symbols.txt'
#	f = open(filename, "r", encoding='utf-8')
#	strings = f.readlines()
	strings = ['Procedure SortArray(n : Integer; Var NQ : Array[1..n] of Real);',
	           'Var k: Real; i,j : Integer;',
	           'Begin for i:=1 to n-1 do']

	st = Stack()
	for s in strings:
		st.push(s)

	strings_obr = []
	while not st.empty():
		s = st.pop()
		print(s)
		strings_obr.append(s)

#	filename = 'symbols_obr.txt'
#	f = open(filename, "w", encoding='utf-8')
#	f.writelines(strings_obr)
#	f.close()


def z9():
	print('___9___')
#	filename = 'logic.txt'
#	f = open(filename, "r", encoding='utf-8')
#	log_str = f.readline()
	log_str = '(((N T) A F) O (N(T X T)))'

	st = Stack()

	def solve(lv, stc):

		def split(l):
			if l[0] == 'T' or l[0] == 'F':
				return l[0], l[1], l[2:]
			else:
				ind = 0
				for i, s in enumerate(l):
					if s == '(':
						ind += 1
					elif s == ')':
						ind -= 1
					if ind == 0:
						return l[:i+1], l[i+1], l[i+2:]

		lv = lv.replace(' ','')
		if lv == 'T':
			stc.push(True)
		elif lv == 'F':
			stc.push(False)
		else:
			lv = lv[1:-1]
			if lv[0] == 'N':
				solve(lv[1:], stc)
				stc.push(not stc.pop())
			else:
				lvsp = split(lv)
				solve(lvsp[2], stc)
				solve(lvsp[0], stc)
				lv_left = stc.pop()
				lv_rigth = stc.pop()
				if lvsp[1] == 'A':
					stc.push(lv_left and lv_rigth)
				elif lvsp[1] == 'X':
					stc.push(lv_left != lv_rigth)
				elif lvsp[1] == 'O':
					stc.push(lv_left or lv_rigth)

	solve(log_str, st)
	print(st.pop())


def z10():
	print('___10___')
#	filename = 'formula.txt'
#	f = open(filename, "r", encoding='utf-8')
#	frm_str = f.readline()
	frm_str = 'M(M(7,N(6,M(M(5,N(4,3)),N(2,1)))),N(0,9))'
	frm_str = frm_str.strip('\n')

	st = Stack()

	def solve(frm, stc):

		def split(f):
			if f[0].isdigit():
				return f[0], f[2:]
			else:
				ind = 0
				for i, s in enumerate(f[1:]):
					if s == '(':
						ind += 1
					elif s == ')':
						ind -= 1
					if ind == 0:
						return f[:i+2], f[i+3:]

		frm = frm.replace(' ','')
		if frm.isdigit():
			stc.push(int(frm))
		else:
			oper = frm[0]
			frm = frm[2:-1]
			frmsp = split(frm)
			solve(frmsp[1], stc)
			solve(frmsp[0], stc)
			frm_left = stc.pop()
			frm_rigth = stc.pop()
			if oper == 'M':
				stc.push(max(frm_left, frm_rigth))
			elif oper == 'N':
				stc.push(min(frm_left, frm_rigth))

	solve(frm_str, st)
	print(st.pop())


def z11():
	print('___11___')
#	filename = 'formula1.txt'
#	f = open(filename, "r", encoding='utf-8')
#	frm_str = f.readline()
	frm_str = '(x + ((y - (z + x)) - (z + x)))'
	frm_str = frm_str.strip('\n')

	st = Stack()

	def solve(frm, stc):

		def isname(s):
			return 'xyz'.find(s) != -1

		def isoper(s):
			return '+-'.find(s) != -1

		def split(f):
			if len(f) < 3:
				return None
			elif isname(f[0]):
				return f[0], f[1], f[2:]
			elif f[0] != '(':
				return None
			else:
				ind = 0
				for i, s in enumerate(f):
					if s == '(':
						ind += 1
					elif s == ')':
						ind -= 1
					if ind < 0:
						return None
					elif ind == 0:
						return f[:i+1], f[i+1], f[i+2:]
				return None

		frm = frm.replace(' ','')
		if isname(frm):
			stc.push(True)
		else:
			frm = frm[1:-1]
			frm_sp = split(frm)
			if frm_sp is None:
				st.push(False)
				return
			solve(frm_sp[2], stc)
			solve(frm_sp[0], stc)
			frm_left = stc.pop()
			frm_rigth = stc.pop()
			st.push(isoper(frm_sp[1]) and frm_left and frm_rigth)

	solve(frm_str, st)
	print(st.pop())


z1()
z2()
z3(4)
z4()
z5()
z6()
z7()
z8()
z9()
z10()
z11()
```
