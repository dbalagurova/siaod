```python
from datetime import datetime
def test_find(func, name):
    
    start_time = datetime.now()
    
    found = func(stroka, str)

    end_time = datetime.now()

    print(name)
    print(found)
    print(end_time - start_time)
    
def simple(main_string, search_string):
    return stroka.index(str)
    
def kmp(main_string, search_string):
    f = [0] * len(search_string)
    for i in range(1, len(search_string)):
        k = f[i-1]
        while k > 0 and search_string[k] != search_string[i]:
            k = f[k-1]
        if search_string[k] == search_string[i]:
            k = k + 1
        f[i] = k

    k = 0
    index = -1
    for i in range(len(main_string)):
        while k > 0 and search_string[k] != main_string[i]:
            k = f[k-1]
        if search_string[k] == main_string[i]:
            k = k + 1
        if k == len(search_string):
            index = i - len(search_string) + 1
            break
    return index


def ubm(main_string, search_string):
    d = [len(search_string)] * 256
    new_p = search_string[::-1]

    for i in range(len(new_p)):
        if d[ord(new_p[i])] != len(new_p):
            continue
        else:
            d[ord(new_p[i])] = i

    len_p = x = j = k = len(search_string)
    counter = 0

    while x <= len(main_string) and j > 0:
        if search_string[j - 1] == main_string[k - 1]:
            j -= 1
            k -= 1
        else:
            x += d[ord(main_string[k - 1])]
            k = x
            j = len_p
            counter += 1

    if j <= 0:
        return k
    else:
        return "Не найдено!"

stroka = input()
str = input()

test_find(simple, "Simple:")
test_find(kmp, "KMP:")
test_find(ubm, "UBM:")
```

```python
import copy


class Position:

	def __init__(self, fields):
		self.fields = copy.deepcopy(fields)

		self.price = 0
		self.calc_position_price()

		pos_zero = self.fields.index(0)
		self.zeroY = pos_zero // 4
		self.zeroX = pos_zero % 4

		self.solvable = False
		self.test_for_solvable()

	@property
	def to_string(self):
		return '{0:2d} {1:2d} {2:2d} {3:2d}\n{4:2d} {5:2d} {6:2d} {7:2d}\n{8:2d} {9:2d} {10:2d} {11:2d}\n{12:2d} {13:2d} {14:2d} {15:2d}'.format(*self.fields)

	def __str__(self):
		return self.to_string

	def __repr__(self):
		return self.to_string

	def calc_position_price(self):
		self.price = 0
		for num, field in enumerate(self.fields, 1):
			if field != 0:
				self.price += (abs(num // 4 - field // 4) + abs(num % 4 - field % 4)) * (2 ** (16 - field))

	def move(self, nx, ny):
		last = 0
		if nx == 0:
			delta = 1 if ny > self.zeroY else -1
			count = self.zeroY - ny + 1 if ny <= self.zeroY else ny - self.zeroY
			for p in range(count):
				new_y = self.zeroY + delta
				self.fields[self.zeroY * 4 + self.zeroX] = self.fields[new_y * 4 + self.zeroX]
				self.fields[new_y * 4 + self.zeroX] = 0
				last = self.fields[self.zeroY * 4 + self.zeroX]
				self.zeroY = new_y
		else:
			delta = 1 if nx > self.zeroX else -1
			count = self.zeroX - nx + 1 if nx <= self.zeroX else nx - self.zeroX
			for p in range(count):
				new_x = self.zeroX + delta
				self.fields[self.zeroY * 4 + self.zeroX] = self.fields[self.zeroY * 4 + new_x]
				self.fields[self.zeroY * 4 + new_x] = 0
				last = self.fields[self.zeroY * 4 + self.zeroX]
				self.zeroX = new_x
		self.calc_position_price()
		return last

	def test_for_solvable(self):
		inv = 0
		for number, elem in enumerate(self.fields):
			if elem != 0:
				for i in range(number):
					if self.fields[i] > elem:
						inv += 1
		inv += self.zeroY + 1
		self.solvable = False
		if inv % 2 == 0:
			self.solvable = True

	def solve(self):
		if not self.solvable:
			return []
		move_variants = ((1, 0), (2, 0), (3, 0), (0, 1), (0, 2), (0, 3))
		last_positions = []
		solve = []
		max_iterations = 1000
		iteration = 0
		while (self.price != 0) and (iteration < max_iterations):
			best_variant = 0
			best_price = 0
			for move_variant in move_variants:
				test_position = Position(self.fields)
				test_position.move(*move_variant)
				if test_position.to_string not in last_positions:
					if best_variant == 0:
						best_variant = move_variant
						best_price = test_position.price
					elif test_position.price < best_price:
						best_variant = move_variant
						best_price = test_position.price
			if best_variant == 0:
				print('Sorry...')
				return []
			solve.append(self.move(*best_variant))
			last_positions.append(self.to_string)
			iteration += 1
			print(self.to_string)
			print('')
		print(solve)


position = Position([1, 2, 3, 4, 5, 6, 7, 8, 13, 9, 11, 12, 10, 14, 15, 0])
print(position)
print(position.solvable)
print('')
position.solve()
```
